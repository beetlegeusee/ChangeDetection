```python
# Python programe to illustrate 
# arithmetic operation of 
# bitwise XOR of two images 
	
# organizing imports 
import cv2 
import numpy as np 
	
# path to input images are specified and 
# images are loaded with imread command 
image1 = imread('pencil.jpg')
image2 = imread('nopencil.jpg')


# cv2.bitwise_xor is applied over the 
# image inputs with applied parameters 
dest_xor = np.bitwise_xor(image1,image2) 

# the window showing output image # with the Bitwise XOR operation 
# on the input images 
imshow( dest_xor) 

# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27: 
	cv2.destroyAllWindows() 
```

```python
from PIL import Image
im = Image.fromarray(dest_xor)
im.save("sample.jpeg")
```

```python
from imageai.Detection import ObjectDetection
import os

execution_path = os.getcwd()

detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath( os.path.join(execution_path , "resnet50_coco_best_v2.0.1.h5"))
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path , "sample.jpeg"), output_image_path=os.path.join(execution_path , "imagenew.jpg"))

for eachObject in detections:
    print(eachObject["name"] , " : " , eachObject["percentage_probability"] )
```
